# Simple Linear Regression Project (Sales)

Modelling the relationship between sales and Advertising

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the dataset
df = pd.read_csv(r"C:\Users\PRISHA\OneDrive\Desktop\SML datasets and files\Assignment 1 SALES.csv", header=None, sep=r"\s+")
df.head()
```

```python
df.shape
# gives shape in the format (rows, columns)
```

```python
df.columns = ["Advertising", "Sales"]
# you couldnt have done this while reading the csv df = df[0].str.split(expand=True) because it was giving error idk why
```

```python
df.head()
```

```python
df.shape
```

```python
# difference between df.describe and df.info
df.info()
```

```python
df.describe()
```

```python
# Declare feature variable and target variable
X = df['Advertising'].values
Y = df['Sales'].values
# When we do this X and Y are arrays there is no concept of row and colunn here
```

```python
# Making a scatter plot
plt.scatter(X, Y, color = 'blue', label = 'Scatter Plot')
plt.title('Relationship between Sales and Advertising')
plt.xlabel('advertising')
plt.ylabel('sales')
plt.legend(loc = 5)
plt.show()
```

```python
# Checking the dimensions of x and y
# We need to check dimensions of x and y to make sure they are in right format for scikit-learn API
# Should x and y have equal number of rows or variable rows are also fine?
# it should have equal number of rows because if you are having an x value there should be a corresponding y
# so we do a sanity check before training you linear regression model and we check if both of them have equal number of values
# if not equal you have to match
print(X.shape)
print(Y.shape)
```

Whenever you have a single dimentional array we as a human we understand that there are 36 rows but your python will not be able to understand if it is 36 rows and single column or single row and 36 columns.
That is the challenge when python works with one dimentional array.
So we have to tell python to take it as one column and 36 rows.
To do this we have to reshape the array.
Reshaping occurs when working with one dimentional arrays

```python
# we use np.array.reshape(-1, 1) which means multiple rows and one column and np.array.reshape(1, -1) means one row multiple columns
# If we dont reshape we get an error
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
```

```python
# Printing the dimentions
print(X.shape)
print(Y.shape)
```

```python
# Split X and Y into training and testing data sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
# Print all training and testing sets and their shapes
print("Shape:", X_train.shape)
print("Shape:", X_test.shape)
print("Shape:", Y_train.shape)
print("Shape:", Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
#Train usin dataset
lm.fit(X_train,Y_train)
#Predict on the test data 
y_pred=lm.predict(X_test)
```

This code showcases Python's object-oriented nature by working with a class (LinearRegression) predefined function , creating an object (lm) ( losses all its properties after run time has its lifetime only till the runtime , and calling methods (fit() and predict()) on that object.
Splits the array in multiple linear regression and have multiple inputs 
Algorithm is not concerend to find simple or continuous linear regression Thus u should know ur data well 

```python
#Computing Slope and intercept 
a=lm.coef_
b=lm.intercept_ 
print("Estimated model slope,a:",a)
print("Estimated model intercept,b:",b)
```

```python
#To make an individual prediction using linear regression 
print((lm.predict([[10]])))#[[]] cause we have it in array if u wanna try to change in to str we use str to typecast 
```

# Regression Metrucs for model performance 
now its the time to evaluate model performance 
We take RMSE cause our ultimate aim is to reduce the value to 0 root ultimately decreases 

```python
#Calculate and preint RMSE
from sklearn.metrics import mean_squared_error 
mse = mean_squared_error(Y_test,y_pred)
rmse = np.sqrt(mse)
print("RMSE value:{: .4f}".format(rmse)) # used a .4 to resrict to 4 decimal places 
#higher R2 better model fits to the data 
```

```python
#Calculate and print r2_score 

from sklearn.metrics import r2_score
print("R2 Score value: {: .4}".format(r2_score(Y_test,y_pred)))
```

```python
#To plot the regression line

plt.scatter(X,Y,color = 'blue' , label="Scatter plot")
plt.plot(X_test,y_pred,color='black',linewidth=3,label = 'Regression line')
plt.title("relationship between sales and advertising")
plt.title("Relationship between Sales and Advertising")
plt.xlabel("Advertising")  # Label for x-axis
plt.ylabel("Sales")  # Label for y-axis
plt.legend()
plt.show()
```

## Simple Linear Regression Project (Car)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\PRISHA\SML LAB\mycar.csv')
df.head()
```

```python
df.shape
```

```python
dF = df.iloc[: , 1:]#drops a specific column 
```

```python
dF.head()
```

```python
dF.info()
```

```python
dF.describe()
```

```python
X = dF['speed'].values
Y = dF['dist'].values
```

```python
dF.columns = ["speed", "dist"]
```

```python
plt.scatter(X, Y, color = 'purple', label = 'Scatter Plot')
plt.title('Relationship between Speed and Distance')
plt.xlabel('speed')
plt.ylabel('dist')
plt.legend(loc = 5)
plt.show()
```

```python
print(X.shape)
print(Y.shape)
```

```python
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
print("Shape:", X_train.shape)
print("Shape:", X_test.shape)
print("Shape:", Y_train.shape)
print("Shape:", Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
#Train usin dataset
lm.fit(X_train,Y_train)
#Predict on the test data 
y_pred=lm.predict(X_test)
```

```python
#Computing Slope and intercept 
a=lm.coef_
b=lm.intercept_ 
print("Estimated model slope,a:",a)
print("Estimated model intercept,b:",b)
```

```python
#Calculate and preint RMSE
from sklearn.metrics import mean_squared_error 
mse = mean_squared_error(Y_test,y_pred)
rmse = np.sqrt(mse)
print("RMSE value:{: .4f}".format(rmse)) # used a .4 to resrict to 4 decimal places 
#higher R2 better model fits to the data 
```

```python
#Calculate and print r2_score 

from sklearn.metrics import r2_score
print("R2 Score value: {: .4}".format(r2_score(Y_test,y_pred)))
```

```python
#To plot the regression line

plt.scatter(X,Y,color = 'blue' , label="Scatter plot")
plt.plot(X_test,y_pred,color='black',linewidth=3,label = 'Regression line')
plt.title("relationship between speed and dist")
plt.xlabel("Advertising")  # Label for x-axis
plt.ylabel("Sales")  # Label for y-axis
plt.legend()
plt.show()
```
