```python
#data set is a form of data structures : all operations performed on array can be performed on dataset 
#mean 
import numpy 
speed=[99,86,87,88,111,86,103,87,94,78,77,85,86]

x=numpy.mean(speed)
y=numpy.median(speed)
#two outcomes from mode operations frequency count and no. with is hghst ocuuring
print("x = ",x)
print("y = ",y)
from scipy import stats#numpy doesnt have mode functions for more extensive stats u use scipy
z=stats.mode(speed)
print(z)
# s.d(sigma) tells about how well the data is spread across if s.d is less the values are close to the avg and if its spread out then its spread across
# entire range 
w = numpy.std(speed)
print("w = ",w)
v = numpy.var(speed)
print("v = ",v)
```
